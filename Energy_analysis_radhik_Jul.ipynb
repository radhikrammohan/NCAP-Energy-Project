Objective is to calcualte the flow rate and enegy variation in a day , week, summer or winter

```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Load the spreadsheet for analysis

```python
# Define the path to your Excel file
input_file_path = 'Jul22 Data.xlsx'

# Define the sheet name
sheet_name = 'Sheet2'

# Read the Excel file
df = pd.read_excel(input_file_path, sheet_name=sheet_name)




```

```python
column_names = df.columns.tolist()

# Print the column names
print("Column Names:", column_names)
```

```python
#unique_values = df['Week'].unique()

Afl_w1 = df[df['Week'] == 1]['field_air_flow.1'].values
Afl_w2 = df[df['Week'] == 2]['field_air_flow.1'].values
Afl_w3 = df[df['Week'] == 3]['field_air_flow.1'].values
Afl_w4 = df[df['Week'] == 4]['field_air_flow.1'].values
Afl_w5 = df[df['Week'] == 5]['field_air_flow.1'].values
Afl_whl= df['field_air_flow.1']



C1_Powerq_w1 = df[df['Week'] == 1]['c1-Power'].values
C1_Powerq_w2= df[df['Week'] == 2]['c1-Power'].values
C1_Powerq_w3 = df[df['Week'] == 3]['c1-Power'].values
C1_Powerq_w4= df[df['Week'] == 4]['c1-Power'].values
C1_Powerq_w5= df[df['Week'] == 5]['c1-Power'].values
C1_Powerq_whl= df['c1-Power']

C2_Power_w1 = df[df['Week'] == 1]['c2-Power'].values
C2_Power_w2 = df[df['Week'] == 2]['c2-Power'].values
C2_Power_w3 = df[df['Week'] == 3]['c2-Power'].values
C2_Power_w4 = df[df['Week'] == 4]['c2-Power'].values
C2_Power_w5 = df[df['Week'] == 5]['c2-Power'].values
C2_Power_whl = df['c2-Power']

C3_Power_w1 = df[df['Week'] == 1]['c3-Power'].values
C3_Power_w2 = df[df['Week'] == 2]['c3-Power'].values
C3_Power_w3 = df[df['Week'] == 3]['c3-Power'].values
C3_Power_w4 = df[df['Week'] == 4]['c3-Power'].values
C3_Power_w5 = df[df['Week'] == 5]['c3-Power'].values
C3_Power_whl = df['c3-Power']

Total_Power_w1 = df[df['Week'] == 1]['Total Power'].values
Total_Power_w2 = df[df['Week'] == 2]['Total Power'].values
Total_Power_w3 = df[df['Week'] == 3]['Total Power'].values
Total_Power_w4 = df[df['Week'] == 4]['Total Power'].values
Total_Power_w5 = df[df['Week'] == 5]['Total Power'].values
Total_Power_whl = df['Total Power']

Th_Power_w1 = df[df['Week'] == 1]['Theoretical Power'].values
Th_Power_w2 = df[df['Week'] == 2]['Theoretical Power'].values
Th_Power_w3 = df[df['Week'] == 3]['Theoretical Power'].values
Th_Power_w4 = df[df['Week'] == 4]['Theoretical Power'].values
Th_Power_w5 = df[df['Week'] == 5]['Theoretical Power'].values
Th_Power_whl = df['Theoretical Power']

Effcy_w1 = df[df['Week'] == 1]['Efficiency'].values
Effcy_w2 = df[df['Week'] == 2]['Efficiency'].values
Effcy_w3= df[df['Week'] == 3]['Efficiency'].values
Effcy_w4 = df[df['Week'] == 4]['Efficiency'].values
Effcy_w5 = df[df['Week'] == 5]['Efficiency'].values
Effcy_whl = df['Efficiency']



```

```python


Eff_weeks =[Effcy_w1,Effcy_w2,Effcy_w3,Effcy_w4]

Max_eff = []
Min_eff = []
Avg_eff = []
Std_eff = []

for week_data in Eff_weeks:
    max_efficiency = np.max(week_data)
    min_eff = np.min(week_data)
    mean_eff =np.mean(week_data)
    std_eff = np.std(week_data)
    Max_eff.append(max_efficiency)
    Min_eff.append(min_eff)
    Avg_eff.append(mean_eff)
    Std_eff.append(std_eff)

#Monthly_data = [Afl_whl,C1_Powerq_whl,C2_Power_whl,C3_Power_whl,Total_Power_whl,Th_Power_whl,Effcy_whl]
Monthly_data = [Effcy_whl]
Monthly_Max = []
Monthly_Min = []
Monthly_Avg= []
Monthly_Std = []

for month_data in Monthly_data:
    max_efficiency = np.max(month_data)
    min_eff = np.min(month_data)
    mean_eff =np.mean(month_data)
    std_eff = np.std(month_data)
    Monthly_Max.append(max_efficiency)
    Monthly_Min.append(min_eff)
    Monthly_Avg.append(mean_eff)
    Monthly_Std.append(std_eff)

print(Monthly_Max)
print(Monthly_Min)
print(Monthly_Avg)
print(Monthly_Std)
```
